```python
import pandas as pd

df = pd.read_csv("../src/data/data_clean.csv")
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

keywords = ["staff", "location", "room", "clean", "service", "wifi"]

data = []
for r in df["clean"]:
    row = {k: (k in r) for k in keywords}
    data.append(row)

df_bin = pd.DataFrame(data)

freq = apriori(df_bin, min_support=0.02, use_colnames=True)
rules = association_rules(freq, metric="confidence", min_threshold=0.5)

rules.sort_values("lift", ascending=False).head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=3000)

df = df.dropna(subset=["clean"])

X = vectorizer.fit_transform(df["clean"])
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=42)
df["cluster"] = kmeans.fit_predict(X)
```
